#### Feature Engineering

```python
def split_location(df_tmp, column):
    # Split the column into new columns
    df_tmp[[f'{column}_City_Place_Code', f'{column}_State']] = df_tmp[column].str.split(' ', expand=True).drop(columns=[2,3,4])

    # Remove the parentheses from 'State'
    df_tmp[f'{column}_State'] = df_tmp[f'{column}_State'].str.replace('(', '').str.replace(')', '')

    # Split the 'City_Place_Code' column into 'City', 'Place', and 'Code'
    df_tmp[[f'{column}_City', f'{column}_Place', f'{column}_Code']] = df_tmp[f'{column}_City_Place_Code'].str.split('_', expand=True).drop(columns=[3])

    # Drop the 'City_Place_Code' column
    df_tmp = df_tmp.drop(columns=[f'{column}_City_Place_Code'])

    return df_tmp
```

```python
#Destination Name: Split and extract features out of destination. City-place-code (State)
df = split_location(df, "destination_name")
#Source Name: Split and extract features out of destination. City-place-code (State)
df = split_location(df, "source_name")
```

```python
#Trip_creation_time: Extract features like month, year and day etc
df['trip_creation_time'] = pd.to_datetime(df['trip_creation_time'])
```

```python
df['trip_year'] = df['trip_creation_time'].dt.year
df['trip_month'] = df['trip_creation_time'].dt.month
df['trip_day'] = df['trip_creation_time'].dt.day
```

#### Indepth Analysis

```python
#Trip Time Taken
# Convert the 'od_start_time' and 'od_end_time' columns to datetime
df['od_start_time'] = pd.to_datetime(df['od_start_time'])
df['od_end_time'] = pd.to_datetime(df['od_end_time'])
# df['trip_time_taken'] = df['od_end_time'] - df['od_start_time']
```

##### Outlier Treatment

```python
#Visualize
def visualize_columns(df):
    # Select numeric columns
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

    # Create a histogram for each numeric column
    for col in numeric_cols:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=df, x=col, kde=True)
        plt.title(f'Histogram of {col}')
        plt.show()
```

```python
visualize_columns(df)
```

```python
import pandas as pd
import numpy as np

def treat_outliers_all_columns(df):
    for column in df.columns:
        if np.issubdtype(df[column].dtype, np.number):  # Only treat numeric columns
            # Calculate IQR
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            # Define limits for outliers
            lower_limit = Q1 - 1.5 * IQR
            upper_limit = Q3 + 1.5 * IQR

            # Treat the outliers
            df[column] = np.where((df[column] < lower_limit), lower_limit, df[column])
            df[column] = np.where((df[column] > upper_limit), upper_limit, df[column])

    return df
```

```python
df = treat_outliers_all_columns(df)
```
