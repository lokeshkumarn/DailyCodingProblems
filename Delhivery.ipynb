#### Feature Engineering

```python
def split_location(df_tmp, column):
    # Split the column into new columns
    df_tmp[[f'{column}_City_Place_Code', f'{column}_State']] = df_tmp[column].str.split(' ', expand=True).drop(columns=[2,3,4])

    # Remove the parentheses from 'State'
    df_tmp[f'{column}_State'] = df_tmp[f'{column}_State'].str.replace('(', '').str.replace(')', '')

    # Split the 'City_Place_Code' column into 'City', 'Place', and 'Code'
    df_tmp[[f'{column}_City', f'{column}_Place', f'{column}_Code']] = df_tmp[f'{column}_City_Place_Code'].str.split('_', expand=True).drop(columns=[3])

    # Drop the 'City_Place_Code' column
    df_tmp = df_tmp.drop(columns=[f'{column}_City_Place_Code'])

    return df_tmp
```

```python
#Destination Name: Split and extract features out of destination. City-place-code (State)
df = split_location(df, "destination_name")
#Source Name: Split and extract features out of destination. City-place-code (State)
df = split_location(df, "source_name")
```

```python
#Trip_creation_time: Extract features like month, year and day etc
df['trip_creation_time'] = pd.to_datetime(df['trip_creation_time'])
```

```python
df['trip_year'] = df['trip_creation_time'].dt.year
df['trip_month'] = df['trip_creation_time'].dt.month
df['trip_day'] = df['trip_creation_time'].dt.day
```
