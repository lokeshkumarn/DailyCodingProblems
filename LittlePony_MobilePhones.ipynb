```python
class Solution:

  def bs_search(self, A, val):

    start = 0
    end = len(A) - 1
    mid = 0

    while start <= end:
      mid = (start + end) // 2
      if A[mid] == val:
        return mid + 1
        #break
      elif A[mid] < val:
        start = mid + 1
      elif A[mid] > val:
        end = mid - 1

    return min(start, end) + 1

  def solve(self, A, B):

        mob_pur = []

        A = sorted(A)

        s = 0
        for a in range(len(A)):
          s = s + A[a]
          A[a] = s


        for m in B:
            count = self.bs_search(A, m)
            mob_pur.append(count)

        return mob_pur     
```
